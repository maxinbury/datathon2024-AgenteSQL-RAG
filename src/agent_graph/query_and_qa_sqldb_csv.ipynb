```python
df = pd.read_csv(here('data/munidataset/siniestros_procesados.csv'))
print(df.shape)
print(df.columns.tolist())
display(df.head(5))
```

```python
from langchain_community.utilities import SQLDatabase
from sqlalchemy import create_engine
db_path = str(here('data')) + '/siniestros_procesados.db'
db_path = f'sqlite:///{db_path}'

engine = create_engine(db_path)
df.to_sql('siniestros_procesados', engine, index=False)
```

```python
db = SQLDatabase(engine=engine)
print(db.dialect)
print(db.get_usable_table_names)
# Asegúrate de que la columna 'cantidad_de_involucrados' existe o ajusta la consulta
try:
    result = db.run('SELECT * FROM siniestros_procesados;')
    print(result)
except Exception as e:
    print(f"Error: {e}")
        
```
